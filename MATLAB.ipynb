# Solving models with the COBRA Toolbox in MATLAB.

The [COBRA toolbox](http://dx.doi.org/doi:10.1038/nprot.2011.308) is a commonly used framework for constraint-based modeling in MATLAB. For more information, please visit its [website](https://opencobra.github.io/cobratoolbox/).

Here we will use [pymatbridge](http://arokem.github.io/python-matlab-bridge/) to run MATLAB scripts which solve all the models in the collection using the COBRA toolbox. They were parsed from SBML/xls into the COBRA toolbox mat format by [this script](http://nbviewer.ipython.org/github/opencobra/m_model_collection/blob/master/convert_models_to_mat.ipynb).

```python
import pandas
from scipy.io import loadmat
```

```python
models = [i for i in sorted(loadmat("all_models.mat").keys()) if not i.startswith("_")]
```

```python
%load_ext pymatbridge
```

Initialize the MATLAB cobra toolbox and solvers. This section may need to be modified to adjust the paths appropriately.

```python
%%matlab --silent
load all_models.mat;
addpath(genpath('~/cobratoolbox/'));
addpath(genpath('/opt/ibm/ILOG/CPLEX_Studio124/cplex/matlab'));
addpath(genpath('/opt/gurobi602/linux64/matlab'));
initCobraToolbox();
```

```python
%%matlab -o r -i models
n_models = size(models, 2);
r = zeros(n_models, 3);

changeCobraSolver('glpk');
for i = 1:n_models
    r(i, 1) = getfield(optimizeCbModel(eval(models{i})), 'f');
end

changeCobraSolver('ibm_cplex');
for i = 1:n_models
    r(i, 2) = getfield(optimizeCbModel(eval(models{i})), 'f');
end

changeCobraSolver('gurobi6');
for i = 1:n_models
    r(i, 3) = getfield(optimizeCbModel(eval(models{i})), 'f');
end
```

```python
rates = pandas.DataFrame(data=r, index=models, columns=["glpk", "cplex", "gurobi"])
rates
```

The maximum difference between these solvers is still less than 1e-6

```python
max(rates.max(axis=1) - rates.min(axis=1))
```
